```python
from PIL import Image

# Charger l'image
image_path = "Datasets/images/train/31.jpeg"  
image = Image.open(image_path)

# Taille de l'image
width_img, height_img = image.size

# Tes coordonnées (pixels)
left = int(width_img * 0.48)
top = int(height_img * 0.62)
right = int(width_img * 0.96)
bottom = int(height_img * 0.96)

# Conversion au format YOLO
x_center = ((left + right) / 2) / width_img
y_center = ((top + bottom) / 2) / height_img
bbox_width = (right - left) / width_img
bbox_height = (bottom - top) / height_img

# Classe (0 = signature dans notre cas)
class_id = 0

# Résultat
yolo_line = f"{class_id} {x_center:.6f} {y_center:.6f} {bbox_width:.6f} {bbox_height:.6f}"
print("YOLO format:", yolo_line)

# Sauvegarde dans un fichier .txt (même nom que l'image)
label_path = image_path.rsplit('.', 1)[0] + ".txt"
with open(label_path, "w") as f:
    f.write(yolo_line + "\n")

print(f"Annotation sauvegardée dans {label_path}")
```

```python
import cv2
from ultralytics import YOLO
import os

# Charger le modèle
model = YOLO("runs/detect/train27/weights/best.pt")

def detect_and_crop(image_path, output_dir="crops"):
    os.makedirs(output_dir, exist_ok=True)
    results = model.predict(image_path, conf=0.5)

    img = cv2.imread(image_path)
    for i, box in enumerate(results[0].boxes.xyxy):  # xmin, ymin, xmax, ymax
        xmin, ymin, xmax, ymax = map(int, box)
        crop = img[ymin:ymax, xmin:xmax]
        cv2.imwrite(f"{output_dir}/signature_{i}.png", crop)
        
# Exemple
detect_and_crop("/home/marco/Projet_du_Fin_DEtude_M2/Datasets/images/test/05.jpg")
```
